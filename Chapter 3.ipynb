# Question 2

Sales of Riding Mowers: Scatter Plots. A company that manufactures riding mowers wants to identify the best sales prospects for an intensive sales campaign. In particular, the
manufacturer is interested in classifying households as prospective owners or nonowners on the basis of Income (in $1000s) and Lot Size (in 1000 ft2). The marketing expert lookedat a random sample of 24 households, given in the file RidingMowers.csv.

a. Using Python, create a scatter plot of Lot Size vs. Income, color-coded by the outcome variable owner/nonowner. Make sure to obtain a well-formatted plot (create legible labels and a legend, etc.).




### Creating a scatter plot of Lot Size vs. Income, color-coded by the outcome variable owner/nonowner, with a well-formatted plot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Importing seaborn for color-coded scatter plot

mower = pd.read_csv('RidingMowers.csv')
mower.head()
```

```python
# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot with color-coded points
sns.scatterplot(x='Lot_Size', y='Income', hue='Ownership', data=mower, palette='Set1', s=100)

# Formatting
plt.title('Scatter Plot of Lot Size vs. Income')
plt.xlabel('Lot Size (in 1000 ft2)')
plt.ylabel('Income (in $1000s)')
plt.grid(True)
plt.legend(title='Ownership')

# Display the plot
plt.tight_layout()
plt.show()
```

# Question 3

Laptop Sales at a London Computer Chain: Bar Charts and Boxplots. The file LaptopSalesJanuary2008.csv contains data for all sales of laptops at a computer chain in London in January 2008. This is a subset of the full dataset that includes
data for the entire year.

a. Create a bar chart, showing the average retail price by store. Which store has the highest average? Which has the lowest?

b. To better compare retail prices across stores, create side-by side boxplots of retail price by store. Now compare the prices in the two stores from (a). Does there seem to be a difference between their price distributions?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

salesjan = pd.read_csv('LaptopSalesJanuary2008.csv')
salesjan.head(2)
```

### Question 3a
#### Creating a bar chart, showing the average retail price by store

```python
# Calculate the average retail price by store
average_prices = salesjan.groupby('Store_Postcode')['Retail Price'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))

# Bar chart of average retail price by store
plt.bar(average_prices['Store_Postcode'], average_prices['Retail Price'], color='b')

# Formatting
plt.title('Average Retail Price by Store')
plt.xlabel('Store')
plt.ylabel('Average Retail Price')
plt.grid(axis='y')

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

## Zooming in to see the stores with the highest  and lowest sales

```python
#Plotting
plt.figure(figsize=(10, 6))

# Bar chart of average retail price by store
plt.bar(average_prices['Store_Postcode'], average_prices['Retail Price'], color='b')

# Formatting
plt.title('Average Retail Price by Store')
plt.xlabel('Store')
plt.ylabel('Average Retail Price')
plt.grid(axis='y')

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')

# # Setting the y-axis limit to zoom in
plt.ylim(480, 500)

# Display the plot
plt.tight_layout()
plt.show()
```

### Question 3b
#### Creating a side-by-side boxplots of retail price by store. Now compare the prices in the two stores

```python
# Side-by-side boxplots using seaborn
sns.boxplot(x='Store_Postcode', y='Retail Price', data=salesjan, palette='Set2')

# Formatting
plt.title('Retail Price Distribution by Store')
plt.xlabel('Store')
plt.ylabel('Retail Price')
plt.grid(axis='y')

# 
# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

# aii. Does price change with time? 
### (Hint: Make sure that the date column is recognized as such. 

The software should then enable different temporal aggregation choices, e.g., plotting the data by weekly or monthly aggregates, or even by day of week.)

```python
# Convert the 'date' column to datetime format
sales['Date'] = pd.to_datetime(sales['Date'])
```

```python
# Set the 'date' column as the index
sales.set_index('Date', inplace=True)

# Resample the data on a weekly basis and sum the values
weekly_data = sales.resample('W').sum()
```

```python
# Plot weekly data
plt.figure(figsize=(12, 6))
sns.lineplot(data=weekly_data, x=weekly_data.index, y='Retail Price', label='Weekly')
plt.title('Weekly Aggregate')
plt.xlabel('Date')
plt.ylabel('Aggregate Value')
plt.legend()
plt.grid()

# Show all x-axis ticks
plt.xticks(rotation=45, ha='right')  # Adjust rotation and ha as needed
plt.show()
```

```python
# Resample the data on a monthly basis and sum the values
monthly_data = sales.resample('M').sum()   

# Plot monthly data
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_data, x=monthly_data.index, y='Retail Price', label='Monthly')
plt.title('Monthly Aggregate')
plt.xlabel('Date')
plt.ylabel('Aggregate Value')
plt.legend()
plt.show()
```
